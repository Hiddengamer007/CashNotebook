```
import subprocess
import sys
import os

def run_setup():
    # 1. Clone the repository (Wait for finish)
    print("--- Cloning Applio ---")
    subprocess.run(["git", "clone", "https://github.com/IAHispano/Applio.git"], check=True)

    # 2. Change directory into the repo
    if os.path.exists("Applio"):
        os.chdir("Applio")
        print(f"Moved to directory: {os.getcwd()}")
    else:
        print("Error: Directory 'Applio' not found after clone.")
        return

    # 3. Prepare and run the installation (Wait for finish)
    # Using check_call to ensure we don't proceed if installation fails
    print("--- Starting Installation (This may take a while) ---")
    subprocess.run(["chmod", "+x", "run-install.sh"], check=True)
    
    # We use 'bash' to run the script directly
    subprocess.run(["bash", "./run-install.sh"], check=True)
    
    print("--- Installation Complete ---")

    # 4. Final Launch (LIVE FEEDBACK MODE)
    print("--- Launching Applio Live ---")
    subprocess.run(["chmod", "+x", "run-applio.sh"], check=True)

    # Launching with Popen for real-time output
    process = subprocess.Popen(
        ["bash", "./run-applio.sh"],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True,
        bufsize=1
    )

    try:
        for line in process.stdout:
            print(line, end='')
            sys.stdout.flush() # Force-push text to the screen
            
            # Watch for the WebUI link
            if "http" in line:
                print("\n" + "="*30)
                print(" WEBUI LINK DETECTED ABOVE ")
                print("="*30 + "\n")
                
    except KeyboardInterrupt:
        print("\nManual shutdown initiated.")
        process.terminate()

    process.wait()

if __name__ == "__main__":
    run_setup()
```
